# 1.Definição do Problema
O presente projeto utiliza o cojunto de dados Climate change impact on agriculture, cujo o arquivo esta disponivel para download e consulta no Kaggel (https://www.kaggle.com/datasets/waqi786/climate-change-impact-on-agriculture).
É um dataset de Aprendizado Supervisionado, especificamente de classificação.
O dataset compreende uma base de dados composto por 15 atributos, sendo que a analise irá concentra-se em um subconjunto dos paises que compoem o BRIC - Brasil, Russia, India e China.

```python
# Configuração para não exibir os warnings
import warnings
warnings.filterwarnings("ignore")

# Importação de pacotes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from IPython.display import display
```

```python
# Informa a url de importação do dataset
URL = "https://raw.githubusercontent.com/Crisvsilva/MVP-Impactos-climaticos-na-agricultura/main/climate_change_impact_on_agriculture_2024.csv"

# Ler o arquivo
Agro = pd.read_csv(URL, delimiter=',')
# Criar um subconjunto com base em uma condição
BRIC = Agro[Agro['Country'].isin(['Brazil','Russia','India','China'])]
```

```python
# Vizualizando as 5 primeiras linhas do dataset para validar a carga
BRIC.head()
```

```python
# Prepara a figura (fig) e a área do gráfico (ax)
fig, ax= plt.subplots(figsize = (10,4))

# Plota o gráfico de barras
ax.bar(x=BRIC['Country'], height=BRIC['Economic_Impact_Million_USD'])

# Exibe o gráfico
plt.show()
```
